```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/shake50_word_topic_distribution.csv')

df.head()
```

```python
result_array = []

# Loop through each topic column
for topic_col in df.columns[1:]:  # Exclude the 'word' column, start from index 1
    topic_index = int(topic_col.split()[1])  # Extract the topic index from the column name
    
    # Find the words with probabilities in the current topic column
    words_with_probs = df[[topic_col, 'word']].sort_values(by=topic_col, ascending=False)
    
    # Find the most representative word with at least 5 letters
    for _, row in words_with_probs.iterrows():
        word = row['word']
        if len(word) >= 1:
            most_representative_word = word
            break  # Exit loop once we find a word with at least 5 letters
    
    # If no word with at least 5 letters is found, choose the word with the highest probability regardless of length
    else:
        most_representative_word = words_with_probs.iloc[0]['word']
    
    result_array.append((topic_index, most_representative_word))  # Append the topic index and the most representative word to the result array


# result_array now contains the most representative word for each topic
print(result_array)
```

```python

import os
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string

def preprocess(text):
    stop = set(stopwords.words('english'))
    exclude = set(string.punctuation)
    lemma = WordNetLemmatizer()
    stop_free = " ".join([word for word in text.lower().split() if word not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
tri = 'This is a text about a e i o u how the hare caught a rabbit, a national geographic documentary. Just to check whether the conceived preprocessor function is working propoerly or not.'

preprocess(tri)
```
